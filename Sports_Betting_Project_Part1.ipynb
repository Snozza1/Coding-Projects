```python
import time
for year in years:
    standings_url="https://www.rugbyleagueproject.org/seasons/nrl-" + str(year) + "/summary.html"
    data = requests.get(standings_url)
    soup = BeautifulSoup(data.text)
    standings_table = soup.select("table.ladder")[0]
    
    links=[l.get("href") for l in standings_table.find_all('a')]
    team_urls = [f"https://www.rugbyleagueproject.org{l}" for l in links]
    
    for team_url in team_urls:
        team_name = team_url.split("/")[-2]
        
        data = requests.get(team_url)
        soup = BeautifulSoup(data.text)
        matches = pd.read_html(data.text)[0]
        matches_cleaned=matches.loc[:, :"Captain(s)"]
        matches_cleaned["Season"] = year
        matches_cleaned["Team"] = team_name
        
        all_matches.append(matches_cleaned)
        time.sleep(1)
```

```python
team_urls
```

```python
match_df=pd.concat(all_matches)
```

```python
match_df.to_csv("matches.csv")
```
